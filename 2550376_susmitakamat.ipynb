1. Create a new column called Happiness Category that categorizes countries into three categories
based on their Score:
Low âˆ’ (Score < 4)
Medium âˆ’ (4 â‰¤ Score â‰¤ 6)
High âˆ’ (Score > 6)

```python
# Categorize countries by happiness score
def happiness_category(score):
    if score < 4:
        return "Low"
    elif 4 <= score <= 6:
        return "Medium"
    else:
        return "High"

# Apply the function to the 'score' column
df['Happiness Category'] = df['score'].apply(happiness_category)

# Display the first 10 rows to verify
df.head(10)

```

Task - 2 - Composite Score Ranking:
Tasks:
1. Using the SouthAsia DataFrame, create a new column called Composite Score that combines the
following metrics:
Composite Score = 0.40 Ã— GDP per Capita + 0.30 Ã— Social Support
+ 0.30 Ã— Healthy Life Expectancy
2. Rank the South Asian countries based on the Composite Score in descending order.
3. Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.
4. Discuss whether the rankings based on the Composite Score align with the original Score - support your

```python
# Step 1: Create new column "Composite Score"
south_asia_df["Composite Score"] = (
    0.40 * south_asia_df["Log GDP per capita"]
    + 0.30 * south_asia_df["Social support"]
    + 0.30 * south_asia_df["Healthy life expectancy"]
)

# Step 2: Rank countries based on Composite Score
south_asia_df = south_asia_df.sort_values(by="Composite Score", ascending=False)

# Step 3: Visualize top 5 countries
top5_composite = south_asia_df.head(5)
plt.figure(figsize=(8,5))
sns.barplot(x="Composite Score", y="Country name", data=top5_composite, palette="plasma")
plt.title("Top 5 South Asian Countries by Composite Score")
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.show()

# Step 4: Compare Composite Score vs Happiness Score
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Composite Score", y="score",
    data=south_asia_df, s=100, color="orange"
)
plt.title("Composite Score vs Happiness Score â€“ South Asia")
plt.xlabel("Composite Score")
plt.ylabel("Happiness Score")
plt.show()
```

Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 Ã— IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# Step 1: Compute IQR for score
Q1 = south_asia_df["score"].quantile(0.25)
Q3 = south_asia_df["score"].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = south_asia_df[(south_asia_df["score"] < lower_bound) | (south_asia_df["score"] > upper_bound)]

print("Outlier Countries based on Happiness Score:")
print(outliers[["Country name", "score", "Log GDP per capita"]])

# Step 3: Scatter plot highlighting outliers
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Log GDP per capita", y="score",
    data=south_asia_df, color="blue", s=100, label="Normal"
)
sns.scatterplot(
    x="Log GDP per capita", y="score",
    data=outliers, color="red", s=120, label="Outlier"
)
plt.title("Outlier Detection â€“ South Asian Countries")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()
```

Task - 5 - Gap Analysis:
Tasks:
1. Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score
for each South Asian country.
2. Rank the South Asian countries by this gap in both ascending and descending order.

â€¢ Calculate the mean, Standard deviation of the score for both South Asia and Middle East.

â€¢ Which region has higher happiness Scores on average?

```python
# Calculate mean and standard deviation of happiness score for both regions
mean_sa = south_asia_df["score"].mean()
std_sa  = south_asia_df["score"].std()
mean_me = middle_east_df["score"].mean()
std_me  = middle_east_df["score"].std()

print("South Asia â€“ Mean:", round(mean_sa, 2), "| Std Dev:", round(std_sa, 2))
print("Middle East â€“ Mean:", round(mean_me, 2), "| Std Dev:", round(std_me, 2))

# Identify which region is happier
if mean_sa > mean_me:
    print("\nSouth Asia has a higher average happiness score.")
else:
    print("\nMiddle East has a higher average happiness score.")
```

â€¢ Identify the top 3 and bottom 3 countries in each region based on the score.

â€¢ Plot bar charts comparing these charts

```python
# Top and bottom 3 for both regions
top3_sa    = south_asia_df.nlargest(3, "score")
bottom3_sa = south_asia_df.nsmallest(3, "score")

top3_me    = middle_east_df.nlargest(3, "score")
bottom3_me = middle_east_df.nsmallest(3, "score")

# Plot comparison
plt.figure(figsize=(9,5))
sns.barplot(x="score", y="Country name", data=top3_sa, color="green", label="South Asia Top 3")
sns.barplot(x="score", y="Country name", data=top3_me, color="orange", label="Middle East Top 3")
plt.title("Top 3 Countries â€“ South Asia vs Middle East")
plt.xlabel("Happiness Score")
plt.legend()
plt.show()
```

â€¢ Compare key metrics like GDP per Capita, Social Support, and Healthy Life Expectancy
between the regions using grouped bar charts.

â€¢ Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.

â€¢ Which region has greater variability in happiness?

```python
# South Asia
range_sa = south_asia_df["score"].max() - south_asia_df["score"].min()
cv_sa    = std_sa / mean_sa

# Middle East
range_me = middle_east_df["score"].max() - middle_east_df["score"].min()
cv_me    = std_me / mean_me

print("South Asia â€“ Range:", round(range_sa,2), "| CV:", round(cv_sa,2))
print("Middle East â€“ Range:", round(range_me,2), "| CV:", round(cv_me,2))

if cv_sa > cv_me:
    print("\nSouth Asia has greater variability in happiness.")
else:
    print("\nMiddle East has greater variability in happiness.")
```

â€¢ Identify outlier countries in both regions based on Score and GDP per Capita.

```python
# ---- South Asia Outlier Detection ----
Q1_sa = south_asia_df['score'].quantile(0.25)
Q3_sa = south_asia_df['score'].quantile(0.75)
IQR_sa = Q3_sa - Q1_sa

low_sa = Q1_sa - 1.5 * IQR_sa
high_sa = Q3_sa + 1.5 * IQR_sa

outliers_sa = south_asia_df[(south_asia_df['score'] < low_sa) | (south_asia_df['score'] > high_sa)]
print("South Asia Outliers:")
print(outliers_sa[['Country name', 'score', 'Log GDP per capita']])

# ---- Middle East Outlier Detection ----
Q1_me = middle_east_df['score'].quantile(0.25)
Q3_me = middle_east_df['score'].quantile(0.75)
IQR_me = Q3_me - Q1_me

low_me = Q1_me - 1.5 * IQR_me
high_me = Q3_me + 1.5 * IQR_me

outliers_me = middle_east_df[(middle_east_df['score'] < low_me) | (middle_east_df['score'] > high_me)]
print("\nMiddle East Outliers:")
print(outliers_me[['Country name', 'score', 'Log GDP per capita']])
```
